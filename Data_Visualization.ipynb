```python
#Q.5 Genarate a synthetic dataset with correlated features.Visualize the correlation matrix of a dataset using a heatmap.
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(42)
size = 100

x1 = np.random.normal(0, 1, size)
x2 = 0.9 * x1 + np.random.normal(0, 0.1, size)
x3 = 0.5 * x1 + np.random.normal(0, 0.2, size)
x4 = np.random.normal(0, 1, size)

data = pd.DataFrame({
    'Feature 1': x1,
    'Feature 2': x2,
    'Feature 3': x3,
    'Feature 4': x4
})

corr_matrix = data.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

z = np.sin(np.sqrt(x**2 + y**2))

data = {'X': x.flatten(), 'Y': y.flatten(), 'Z': z.flatten()}
df = pd.DataFrame(data)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')

ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('3D Surface Plot of z = sin(√(x² + y²))')

plt.show()
```

```python
import numpy as np
from bokeh.plotting import figure, show, output_notebook

x = np.linspace(0, 10, 500)
y = np.sin(x)

p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)', width=700, height=400)

p.line(x, y, legend_label="sin(x)", line_width=2)

output_notebook()

show(p)
```

```python
#Q.2 Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes'and'colors'columns.
```

```python
import numpy as np
from bokeh.plotting import figure, show, output_notebook
from bokeh.io import push_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColumnDataSource

n_points = 500
x = np.random.rand(n_points) * 100
y = np.random.rand(n_points) * 100
sizes = np.random.rand(n_points) * 30
colors = np.random.rand(n_points) * 100

source = ColumnDataSource(data=dict(x=x, y=y, sizes=sizes, colors=colors))

p = figure(title="Random Scatter Plot", x_axis_label='x', y_axis_label='y', width=700, height=400)

p.scatter('x', 'y', size='sizes', color=linear_cmap('colors', 'Viridis256', 0, 100), source=source, fill_alpha=0.6)

output_notebook()

show(p)
```

```python
#Q.3 Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.fruits = ['Apples','Oranges','Bananas','Pears'] counts = [20,25,30,35]
```

```python
from bokeh.plotting import figure, show, output_notebook
from bokeh.io import push_notebook
from bokeh.transform import factor_cmap
from bokeh.models import ColumnDataSource

fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

p = figure(x_range=fruits, title="Fruit Counts", x_axis_label='Fruit', y_axis_label='Count', width=700, height=400)

p.vbar(x='fruits', top='counts', width=0.5, source=source, legend_label="Fruits",
       line_color='white', fill_color=factor_cmap('fruits', palette=['#c9d9d3', '#718dbf', '#e84d60', '#ddb7b1'], factors=fruits))

p.legend.orientation = "horizontal"
p.legend.location = "top_center"

output_notebook()

show(p)
```

```python
#Q.4 Create a Bokeh histogram to visualize the distribution of the given data.data_hist = np.random.randn(1000)hist'edges = np.histogram(data_hist'bins=30)
```

```python
import numpy as np
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColumnDataSource

data_hist = np.random.randn(1000)

hist, edges = np.histogram(data_hist, bins=30)

source = ColumnDataSource(data=dict(top=hist, left=edges[:-1], right=edges[1:]))

p = figure(title="Histogram of Random Data", x_axis_label='Value', y_axis_label='Frequency', width=700, height=400)

p.quad(top='top', bottom=0, left='left', right='right', source=source, fill_color="green", line_color="white", alpha=0.6)

output_notebook()

show(p)
```

```python
#Q.5 Create a bokeh heatmap using the provided dataset.data_heatmap = np.random.rand(10, 10) x = np.linspace(0,1,10) y = np.linspace(0,1,10) xx, yy = np.meshgrid(x,y)
```

```python
import numpy as np
from bokeh.plotting import figure, show, output_notebook
from bokeh.models import ColorBar, LinearColorMapper
from bokeh.transform import linear_cmap
from bokeh.io import push_notebook
from bokeh.layouts import layout

data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

p = figure(title="Heatmap", x_range=(0, 1), y_range=(0, 1), x_axis_label='X', y_axis_label='Y', width=700, height=700)

color_mapper = LinearColorMapper(palette="Viridis256", low=np.min(data_heatmap), high=np.max(data_heatmap))

p.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1, color_mapper=color_mapper)

color_bar = ColorBar(color_mapper=color_mapper, location=(0, 0))
p.add_layout(color_bar, 'right')

output_notebook()

show(p)
```
